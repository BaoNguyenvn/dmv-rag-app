```python
import os
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS
from dotenv import load_dotenv
```

```python

load_dotenv()

# Now you can access it like this:
api_key = os.getenv("OPENAI_API_KEY")
```

```python
documents = []
filename = r'D:\Machine Learning Semester 2\Ragassignment\DMV.pdf'
loader= PyPDFLoader(filename)
documents.extend(loader.load())
documents
print(f"Loaded {len(documents)} pages from PDFs")
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,      # ~250 words per chunk
    chunk_overlap=50     # Keep some context between chunks
)

#break down into smaller pieces (since models have limit capacities).
chunks = text_splitter.split_documents(documents)
print(f"Split into {len(chunks)} chunks")
print(chunks[:1])
```

```python
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")  # QUAN TRỌNG
vectorstore = FAISS.from_documents(chunks, embeddings)
vectorstore.save_local("faiss_index")
print("Done! faiss_index saved.")
```
